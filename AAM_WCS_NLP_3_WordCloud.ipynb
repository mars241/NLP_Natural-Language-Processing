1. Un WordCloud √† partir du texte initial, gr√¢ce √† la m√©thode **generate_from_text**

2. Un WordCloud gr√¢ce √† la m√©thode **generate_from_frequencies** apr√®s nettoyage des stopwords et de la ponctuation.

```python
# Traitement du text originel (nettoyage des stopwords et de la ponctuation) et le sortir en liste de mot:
sentence = ('This is a story about dogs our canine pets Dogs are furry animals ! Ceci est une histoire de chiens nos animaux de compagnie. Les chiens sont des animaux √† fourrure !')
sentence = sentence.lower() #Mettre minuscule


import string # Importer le module correspondant
#string.punctuation #  Pour afficher la liste des ponctuations vis√©s
# Enlever la ponctuation
for punctuation in string.punctuation:
    sentence = sentence.replace(punctuation, ' ')

# T√©l√©chargement du pakage stop word
nltk.download('stopwords')
# La biblioth√®que des stops words en Anglais
from nltk.corpus import stopwords
# La biblioth√®que de la tozennisation
from nltk.tokenize import word_tokenize

# Pour enlever les stops words des tokens en Anglais
tokens = nltk.word_tokenize(sentence)
tokens_clean = []
for words in tokens:
  if words not in nltk.corpus.stopwords.words("english"):
    tokens_clean.append(words)

tokens_clean
```

```python
# Cr√©er le "WordCloud" apr√®s nettoyage des stopwords et de la ponctuation( √† partir d'une liste de mots)
# Import du module de "WordCloud"
from wordcloud import WordCloud
# G√©n√©ration du wordCloud depuis une liste de mots avec la m√©thode "generate_from_text" (joindre en text au pr√©alable avec ".join")
string_tokens_clean = ' '.join(tokens_clean)
wordcloud.generate_from_text(string_tokens_clean)

# Affichage gr√¢ce √† Matplotlib
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

3. Un WordCloud gr√¢ce √† la m√©thode **generate_from_frequencies** apr√®s nettoyage des stopwords et de la ponctuation, et application d'un **stemmer.**

```python
# Il est rare la g√©n√©ration d'un "WordCloud" directe depuis un texte, car nous pr√©f√©rons cr√©er un wordcloud apr√®s du processing.
# Nous avons donc souvent un dictionnaire de FreqDist.

# importer le module pour le "stemmer"
from nltk.stem import SnowballStemmer
# Nous pouvons maintenant faire un FreqDist sur cette nouvelle liste de tokens de la phrase compl√®te :
full_vocab = nltk.FreqDist(tokens_clean)
full_vocab # ici full_vocab est le dictionnaire(FreqDist)

# importer le "wordcloud"
from wordcloud import WordCloud
# G√©n√©ration du wordCloud depuis un texte avec la m√©thode "generate_from_frequencies" avec dictionnaire
wordcloud = wordcloud.generate_from_frequencies(full_vocab)

# Affichage gr√¢ce √† Matplotlib
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

4) Un WordCloud gr√¢ce √† la m√©thode **generate_from_frequencies** apr√®s nettoyage des stopwords et de la ponctuation, et application d'un **lemmatizer**. Partage le lien vers ton notebook en guise de solution au challenge.

```python
# traitrement du text originel (nettoyage des stopwords et de la ponctuation et application d'un lemmatizer)

sentence = ('This is a story about dogs our canine pets Dogs are furry animals ! Ceci est une histoire de chiens nos animaux de compagnie. Les chiens sont des animaux √† fourrure !')
import string # Importer le module correspondant
# string.punctuation #  Pour afficher la liste des ponctuations vis√©s
# Enlever la ponctuation
for punctuation in string.punctuation:
    sentence = sentence.replace(punctuation, ' ')

# import de la biblioth√®que de tokenisation
from nltk.tokenize import word_tokenize
# tokenisation
# Pour enlever les stops words des tokens en Anglais
tokens = nltk.word_tokenize(sentence)
tokens_clean = [] # initialisation
for words in tokens:
  if words not in nltk.corpus.stopwords.words("english"):
    tokens_clean.append(words)
# ici "tokens_clean" est la liste de la "sentence" tokeniz√©, elle est remplie(.append) de tout les mots

# joindre en string la liste des mots toz√©nis√©s:
string_mots = ' '.join(tokens_clean)

#Lemmatizeration
import spacy
# Import des r√®gles anglaises <=> correspond un "import spacy as nlp"
nlp = spacy.load('en_core_web_sm')

# spacy d√©coupe automatiquement en tokens avec cette syntaxe :
sent_tokens = nlp(string_mots)

list_lemma = [] # initialisation
for token in sent_tokens:
  list_lemma.append(token.lemma_)
  print(token, token.lemma_)
```

```python
# On applique les stop word √† la liste Lemmatiz√©e
mots_clean_stop = [i for i in list_lemma if not i in nltk.corpus.stopwords.words("english")]
# On reconstruit le texte avec les mots tozenis√©s et Lemmatiz√©e:
paragraphe_clean_stop = ' '.join(mots_clean_stop)
paragraphe_clean_stop
```

```python
## Affichage du "worldcloud" ##

# import de la biblioth√®que du lemmatizer(spacy)
import spacy
# Import des r√®gles anglaises
nlp = spacy.load('en_core_web_sm')


# Ici, nous repartons de notre lemmatizer spacy :

sent_tokens = nlp(paragraphe_clean_stop)

# / ! \ ici pour utiliser la m√©thode "generate_from_frequencies" avec un texte ("paragraphe_clean_stop") est caduc, il nous faut un dictionaire
# D'o√π la question: comment convertir le texte lemmatiz√© en dictionaire sans repartir du dictionaire de base fait avec le "FreqDist" ;> full_vocab = nltk.FreqDist(tokens_clean)
# D'autant que dans cette √©tape on a perdu la comptabilisation des mots(c_√†_d les valeurs(= fr√©quence des mots) des cl√©s(=mots))!

# Autre mani√®re de "lemmatiz√©"
from nltk.stem import SnowballStemmer
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Instancier le stemmer Snowball pour l'anglais
stemmer = SnowballStemmer("english")

#tokenisation
words = nltk.word_tokenize(paragraphe_clean_stop)
full_vocab_stemmed = [stemmer.stem(word) for word in words]  #stemming des mots

# Ici stemmage: r√©duire les mots √† leur racine et le ".join()" les regroupe en liste(une s√©quence modifiable de donn√©es ordonn√©es); donc on peut les comptabilis√©s
stemmed_string = " ".join(full_vocab_stemmed)
# Cr√©ation du dictionnaire de fr√©quences
stemmed_freq_dist = nltk.FreqDist(full_vocab_stemmed)
# Affichage du "stemmed_freq_dist" qui est le DICO_LAMMENIZER_DES_FREQUENCES_A_METTRE DANS LE "generate_from_frequencies()"
stemmed_freq_dist

# Remarque:  Le dico semble dans nos attentes, prochaine √©tape g√©n√©ration du "WordCloud" !
```

```python
# importer le "wordcloud"
from wordcloud import WordCloud # (A NOTER : la diff√©rence entre "wordcloud" & "WordCloud" !)

# Param√®trage du "wordcloud" √† l'aide de la fct ".generate_from_frequencies"
wordcloud = wordcloud.generate_from_frequencies(stemmed_freq_dist)

# Affichage gr√¢ce √† Matplotlib
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(wordcloud , interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
