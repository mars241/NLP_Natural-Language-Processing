# Pr√©traitement de texte

Le pr√©traitement des textes est naturellement diff√©rent du pr√©traitement num√©rique classique. Cependant, il est tout aussi important. Les t√¢ches courantes de pr√©traitement sont les suivantes :

* lowercase
* Traitement des chiffres et de la ponctuation
* Suppression des "Stopwords".
* Tokenisation
* D√©doublement ou lemmatisation

## LowerCase

Les algorithmes de mod√©lisation de texte sont sensibles √† la diff√©rence entre minuscule et majuscule.

```python
text = '999 Football is my passion. I love to play with footballs!! Who else loves football?'

text = text.lower()
text
```

## Les nombres

En fonction de la t√¢che, il peut √™tre n√©cessaire de supprimer les num√©ros dans le cadre du pr√©traitement.

```python
text = ''.join(word for word in text if not word.isdigit())

text
```

üëç Les nombres sont utiles pour l'extraction de dates

üëé Les nombres sont inutiles pour la mod√©lisation de certains sujets.

# Ponctuation

Comme les chiffres, la ponctuation peut devoir √™tre supprim√©e dans le cadre du pr√©traitement.

```python
import string

string.punctuation
```

```python
for punctuation in string.punctuation:
    text = text.replace(punctuation, ' ')

text
```

‚ö†Ô∏è La ponctuation est rarement respect√©e dans les formes de texte modernes (par exemple, les m√©dias sociaux).

# Vectorisation

Les algorithmes d'apprentissage automatique ne peuvent pas traiter du texte brut. Le processus de transformation du texte brut en une repr√©sentation num√©rique est appel√© vectorisation. Il existe trois techniques de vectorisation standard :

* "Bag of words"
* Tf-Idf
* N-grams

### Bag of words
La repr√©sentation par sac de mots consiste √† compter les occurrences de chaque mot dans un texte.  
Le nombre d'occurrences de chaque mot devient une caract√©ristique.



Capture d‚ÄôeÃÅcran 2021-10-14 aÃÄ 16.38.25.png

`CountVectorizer` de Sklearn
Un outil pour g√©n√©rer automatiquement une repr√©sentation en "Bag of Words" d'un texte.


```python
texts = ['i love football',
         'football is a game i love',
        'football football football']

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(texts)

X.toarray()
```

```python
vectorizer.get_feature_names()
```

```python
texts
```

```python
import pandas as pd

pd.DataFrame(X.toarray(),columns = vectorizer.get_feature_names())
```

Limites de la repr√©sentation en "Bag of Words":

* Ne prend pas en compte la longueur du document
* Ne tient pas compte le contexte

# Tf-Idf
`Term Frequency - Inverse Document Frequency` note les mots selon leur importance dans un texte, en fonction de leur pr√©sence dans une collection de documents.

TF-IDF d√©finit l'importance d'un terme en prenant en consid√©ration l'importance de ce terme dans un document unique et en l' √©chelonnant en fonction de son importance dans tous les documents.

https://towardsdatascience.com/how-tf-idf-works-3dbf35e568f0

## TfidfVectorizer
Un outil pour g√©n√©rer automatiquement une repr√©sentation Tf-idf d'un texte:

```python
from sklearn.feature_extraction.text import TfidfVectorizer

texts = ['i love football',
         'football is a game i love',
        'football football football']

tf_idf_vectorizer = TfidfVectorizer()

X = tf_idf_vectorizer.fit_transform(texts)

X.toarray()
```

```python
pd.DataFrame(X.toarray(),columns = tf_idf_vectorizer.get_feature_names())
```

üëç Avantages de la repr√©sentation Tf-Idf :

* L'utilisation de la fr√©quence plut√¥t que du comptage est robuste √† la longueur du document.

üëé Inconv√©nient de la repr√©sentation Tf-Idf :

* Ne capture pas le contexte

#### Hyperparam√®tres cl√©s de `CountVectorizer` et de `TfidfVectorizer`.
* `max_df`
* `min_df`
* `max_features`

### `max_df`

Utilis√© pour exclure les "stopwords sp√©cifiques au corpus", c'est-√†-dire les mots qui sont tr√®s fr√©quents dans l'ensemble de donn√©es.  
Le vectoriseur ignorera les mots dont la fr√©quence est sup√©rieure au seuil sp√©cifi√©.

```python
texts
```

```python
tf_idf_vectorizer = TfidfVectorizer(max_df = 0.67)

X = tf_idf_vectorizer.fit_transform(texts)

X.toarray()

pd.DataFrame(X.toarray(),columns = tf_idf_vectorizer.get_feature_names())
```

üëâ Particuli√®rement utile pour supprimer les mots qui sont si fr√©quents qu'ils ont peu de pouvoir pr√©dictif.

*Exemple* : Lors de la classification de textes en th√®mes Basket-ball ou football, le mot "ballon" appara√Ætra souvent, mais ne sera pas utile pour pr√©dire l'un ou l'autre.

### `min_df`

Utilis√© pour exclure les mots qui sont tr√®s peu fr√©quents dans l'ensemble de donn√©es. Le vectoriseur ignorera les mots dont la fr√©quence est inf√©rieure au seuil sp√©cifi√©.

```python
texts
```

```python
tf_idf_vectorizer = TfidfVectorizer(min_df = 0.5)

X = tf_idf_vectorizer.fit_transform(texts)

X.toarray()

pd.DataFrame(X.toarray(),columns = tf_idf_vectorizer.get_feature_names())
```

üëâ Particuli√®rement utile pour supprimer les fautes de frappe ou les anomalies du texte manqu√©es lors du pr√©traitement.

### `max_features`

Utilis√© pour sp√©cifier le nombre de caract√©ristiques √† conserver lors de la vectorisation. Il retiendra les caract√©ristiques les plus importantes selon le nombre ou le score tf-idf.

```python
tf_idf_vectorizer = TfidfVectorizer(max_features = 2)

X = tf_idf_vectorizer.fit_transform(texts)

X.toarray()

pd.DataFrame(X.toarray(),columns = tf_idf_vectorizer.get_feature_names())
```

üëâ Particuli√®rement utile pour r√©duire la dimension des donn√©es.

# N-Gram
Au lieu de consid√©rer des mots individuels, les "N-gram" consistent √† consid√©rer des s√©quences de mots. Cette repr√©sentation permet de capturer le contexte. N est le nombre de mots √† consid√©rer.

### L' hyperparam√®tre `ngram_range`
Permet de sp√©cifier la longueur des s√©quences √† consid√©rer.

```python
texts
```

```python
tf_idf_vectorizer = TfidfVectorizer(ngram_range = (1,3))

X = tf_idf_vectorizer.fit_transform(texts)

X.toarray()

pd.DataFrame(X.toarray(),columns = tf_idf_vectorizer.get_feature_names())
```

# La richesse de vocabulaire

```python
data = ["i do not love football",
        "i love football not basketball",
        "football football football"]

data_df = pd.DataFrame(data)
data_df.columns = ["Texte"]

data_df
```

```python
def vocab_richness(text):
    tokens = word_tokenize(text)
    total_length = len(tokens)
    unique_words = set(tokens)
    print(unique_words)
    unique_word_length = len(unique_words)
    return unique_word_length/total_length

data_df['vocab richness'] = data_df["Texte"].apply(vocab_richness)

data_df
```
