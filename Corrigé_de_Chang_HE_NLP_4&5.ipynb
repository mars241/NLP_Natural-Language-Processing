```python
#  importation de la Bibliothèque pour la lecture/traitement de DF (train.csv)
import pandas as pd
# importation du dé-zippeur
import zipfile

#df = pd.read_csv("/Users/changerer/Desktop/DataAnalyst/tweets.zip", compression='zip')

# My link :
# link_zip_folder = "C:/Users/marcia2/Desktop/Data/tweets.zip"
# df = pd.read_csv(link_zip_folder, compression='zip')

zip_name='tweets.zip'
csv_name='train.csv'
myzip = zipfile.ZipFile(zip_name) # open zip  # ouvrir le zip
file = myzip.open(csv_name) # open csv # ouvrir le  csv

df= pd.read_csv(file) # pandas read csv # pandas pour lire le csv
df.head()
```

```python
df = pd.read_csv("tweets.zip", compression='zip')
df.info()
```

```python
# Ne garde que les tweets positifs et négatifs (tu excluras donc les neutral). Quel est le pourcentage de tweets positifs/négatifs ?
pourcentage = (1 - len(df[df['sentiment'] == 'neutral'])/len(df)) * 100
print(round(pourcentage, 2),'%')
```

```python
df = df[df['sentiment'] != 'neutral']
```

```python
# Copie la colonne text dans une Série X, et la colonne sentiment dans une Série y.
# Applique un train test split avec le random_state = 32 et un train_size de 0.75.

from sklearn.model_selection import train_test_split
X = df['text']
y = df['sentiment']
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.75, random_state=32)
```

# Quête 4 :
### Crée un modèle vectorizer avec scikit-learn en utilisant la méthode Countvectorizer.  Entraîne ton modèle sur X_train, puis crée une matrice de features X_train_CV. Crée la matrice X_test_CV sans ré-entraîner le modèle. Le format de la matrice X_test_CV doit être 4091x15806 avec 44633 stored elements.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train_CV = vectorizer.fit_transform(X_train)
X_train_CV
```

```python
vectorizer.get_feature_names_out()
```

```python
pd.DataFrame(X_train_CV.toarray(),columns = vectorizer.get_feature_names_out())
```

```python
X_train_CV = vectorizer.transform(X_train)
X_train_CV
```

```python
X_test_CV = vectorizer.transform(X_test)
X_test_CV
```



### 1.   Entraîne maintenant une régression logistique avec les paramètres par défaut. Tu devrais obtenir les résultats suivants : 0.966 pour le test d'entraînement, et 0.877 pour l'ensemble de test.
### 2.   Étape bonus : essaye d'afficher 10 tweets qui ont été mal prédits (faux positifs ou faux négatifs). Aurais-tu fait mieux que l'algorithme ?



```python
# régression logistique

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_CV,y_train)
print("Accuracy score sur l'ensemble d'entraînement :", '%.3f'%model.score(X_train_CV, y_train))
print("Accuracy score sur l'ensemble de test :", '%.3f'%model.score(X_test_CV, y_test))
```

```python
y_predicted = model.predict(X_test_CV)
for prediction, truth in zip(y_predicted[:10], y_test[:10]):
    print(prediction, truth)
```

```python
# Print the Confusion Matrix on the test set
from sklearn.metrics import confusion_matrix
pd.DataFrame(data= confusion_matrix(y_true = y_test, y_pred = y_predicted),
             index = model.classes_ + " TRUTH",
             columns = model.classes_ + " PREDICTED")
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_fscore_support


print(classification_report(y_test, y_predicted))
```

### Crée un modèle vectorizer avec scikit-learn en utilisant la méthode TfidfVectorizer. Entraîne ton modèle sur X_train, puis crée une matrice de features X_train_CV. Crée la matrice X_test_CV sans ré-entraîner le modèle. Le format de la matrice X_test_CV doit être 4091x15806 avec 44633 stored elements.
### Entraîne maintenant une régression logistique avec les paramètres par défaut. Tu devrais obtenir les résultats suivants : 0.932 pour le test d'entraînement, et 0.873 pour l'ensemble de test.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train)
X_train_tfidf
```

```python
pd.DataFrame(X_train_tfidf.toarray(),columns = tfidf.get_feature_names_out())
```

```python
X_test_tfidf = tfidf.transform(X_test)
X_test_tfidf
```

```python
pd.DataFrame(X_test_tfidf.toarray(),columns = tfidf.get_feature_names_out())
```

```python
# régression logistique

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_tfidf,y_train)
print("Accuracy score sur l'ensemble d'entraînement :", '%.3f'%model.score(X_train_tfidf, y_train))
print("Accuracy score sur l'ensemble de test :", '%.3f'%model.score(X_test_tfidf, y_test))
```
